```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
```

```python
data = pd.read_csv("Language Detection.csv")
```

```python
X = data["Text"]
y = data["Language"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# creating a list for appending the preprocessed text
data_list = []
# iterating through all the text
for text in X:
       # removing the symbols and numbers
        text = re.sub(r'[!@#$(),n"%^*?:;~`0-9]', ' ', text)
        text = re.sub(r'[[]]', ' ', text)
        # converting the text to lower case
        text = text.lower()
        # appending to data_list
        data_list.append(text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(data_list).toarray()
X.shape # (10337, 39419)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
ac = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy is :",ac)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(cm, annot = True)
plt.show()
```

```python
def predict(text):
     x = cv.transform([text]).toarray() # converting text to bag of words model (Vector)
     lang = model.predict(x) # predicting the language
     lang = le.inverse_transform(lang) # finding the language corresponding the the predicted value
     print("The langauge is in",lang[0]) # printing the language
```

```python
predict("Welcome to sharad mittal's blog")
```

```python
predict("मित्तल के ब्लॉग पर आपका स्वागत है")
```

```python
predict("Καλώς ήλθατε στο blog του sharad mittal")
```

## Pickling the model

```python
import pickle
```

```python
#pickling the model
filename = "language_predictor.pkl"
pickle.dump(model, open(filename, 'wb'))
```

```python
# pickling the count vectorizer
count_vect = "vectorize.pkl"
pickle.dump(cv, open(count_vect, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
def load_predict(text):
     x = cv.transform([text]).toarray() # converting text to bag of words model (Vector)
     lang = loaded_model.predict(x) # predicting the language
     lang = le.inverse_transform(lang) # finding the language corresponding the the predicted value
     print("The langauge is in",lang[0]) # printing the language
```

```python
load_predict("Welcome to sharad mittal's blog")
```
