```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('data/processed/en/train.pkl', 'rb') as f:
    data = pickle.load(f)
data_len = [len(sentence) for sentence in data]
```

```python
plt.hist(data_len, bins=96, range=[0, 96])
plt.xlabel('Number of Tokens')
plt.ylabel('Amount of Sentences')
plt.title('Number of Tokens in Each Sentence')
plt.savefig('output/data.png')
plt.show()
```
